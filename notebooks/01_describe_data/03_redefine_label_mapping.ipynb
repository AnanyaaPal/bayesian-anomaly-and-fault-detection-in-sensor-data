```python
import pandas as pd
import numpy as np

import path
from path import Path
```

## Load (Raw & Unmapped) Train Labels

```python
dir_root = Path.cwd().parent.parent
dir_root
```

```python
dir_train_labels = dir_root / "data" / "train" / "y_train.txt"
dir_train_labels
```

```python
train_labels_unmapped = pd.read_csv(dir_train_labels,
                                    header=None,
                                    index_col=None
                                    )
display(f"number of observations in the training data: {len(train_labels_unmapped)}")
display(f"confirming shape of input labels: {train_labels_unmapped.shape}")
```

Count existing labels to confirm the new mapping later.

```python
label_wise_sum = train_labels_unmapped[0].value_counts().sort_index()
label_wise_sum
```

```python
label_wise_sum.shape
```

```python
initial_sum_label_0 = label_wise_sum.iloc[0:3, ].sum()
initial_sum_label_0
```

```python
initial_sum_label_1 = label_wise_sum.iloc[3:6, ].sum()
initial_sum_label_1
```

## Mapping existing class labels into 2 classes to prepare data for binary classification using Na√Øve Bayes

- Map current labels [1,2,3] to label 0 : Dynamic
- Map current labels [4,5,6] to label 1: Static

```python
label_map = {
    1: 0, 2: 0, 3: 0,
    4: 1, 5: 1, 6: 1
}
train_labels_mapped = train_labels_unmapped.copy()

train_labels_mapped = train_labels_unmapped[0].map(label_map)
train_labels_mapped
```

```python
sum_of_new_label_mappings = train_labels_mapped.value_counts().sort_index()
sum_of_new_label_mappings
```

```python
sum_of_new_label_mappings.shape
```

Confirm whether mapping matches original class label distribution.

```python
if sum_of_new_label_mappings[0] == initial_sum_label_0:
    print("New label mappings for Class 0: Dynamic are correct.")

if sum_of_new_label_mappings[1] == initial_sum_label_1:
    print("New label mappings for Class 1: Static are correct.")
```

## Save new label mappings for train data.

```python
len(train_labels_mapped)
```
